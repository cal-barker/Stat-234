```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
```

```python
# URL
url = "https://och.byu.edu/all-facilities"
html = urlopen(url)
soup = BeautifulSoup(html)

df = pd.DataFrame(columns = ['value'])
```

```python
soup
#soup.findAll('tr', limit=2)
results = soup.findAll('td', class_='line1')


for i in range(0, len(results)):
    string = str(results[i])
    string = string.replace('"', "").replace("<td class=line1>", "").replace("</td>","")
    df = df.append({'value' : string}, ignore_index = True)
#df.head()
```

```python
searchfor = [' N ', ' S ', ' E ', ' W', '\n', 'hide', 'Large', 'Small', 'Medium', 'Condominium']
df= df[df["value"].str.contains('|'.join(searchfor)) == False]
```

```python
df_true = df[df['value'] != ""]
df_true = df_true.drop_duplicates() 
```

```python
df_true.to_csv("~/Desktop/housing.csv")
```
